```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def unit_step(t):
    return np.heaviside(t,1)
```

```python
t=np.linspace(-5,5,1000)
y=unit_step(t)
```

```python
plt.plot(t,y,'r')
plt.title("Continous time Unit step plot")
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid()
plt.show()
```

```python
t1=np.linspace(-5,5,40)
y1=unit_step(t1)
plt.stem(t1,y1,'k')
plt.title('discrete time unit step plot')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid()
plt.show()
```
